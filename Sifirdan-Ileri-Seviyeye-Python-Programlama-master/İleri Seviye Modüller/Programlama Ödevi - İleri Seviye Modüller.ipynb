### Proje 1

Bilgisayarınızdaki tüm mp4,txt ve pdf dosyalarını **os** modülüyle arayın ve bunların nerede bulunduklarını ve isimlerini ayrı ayrı "pdf_dosyalari.txt","mp4_dosyaları.txt","txt_dosyaları.txt" adlı dosyalara kaydedin.

```python
import os
path = "C:\\Users\\HP\\Desktop"
files = []
roots = []
directories = []
for (r,d,f) in os.walk(path):
    for i in f:
        files.append(i)
    for i in r:
        roots.append(i)
    for i in d:
        directories.append(i)
    

filemp = "Mp4 Dosyaları.txt"
summp = 0
with open(os.path.join(path, filemp), 'w') as fmp:
    for i in files:
        if(i.endswith(".mp4")):
            summp+=1
            fmp.write(f"{summp}. Dosya = {i}" + "\n")
            #print(i)

filetxt = "Txt Dosyaları.txt"
sumtxt =0
with open(os.path.join(path, filetxt), 'w') as ftxt:
    for i in files:
        if(i.endswith(".txt")):
            sumtxt+=1
            ftxt.write(f"{sumtxt}. Dosya = {i}" + "\n")
            #print(i)

filepdf = "Pdf Dosyaları.txt"
sumpdf =0
with open(os.path.join(path, filepdf), 'w') as fpdf:
    for i in files:
        if(i.endswith(".pdf")):
            sumpdf+=1
            fpdf.write(f"{sumpdf}. Dosya = {i}" + "\n")
            #print(i)





print(f"MP4:{summp}\nTXT:{sumtxt}\nPDF:{sumpdf}")
```
